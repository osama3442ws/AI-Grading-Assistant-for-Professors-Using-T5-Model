```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import PyPDF2
import gradio as gr
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import os

# Load T5 model from Hugging Face
model_name = "t5-small"  # You can try other models as well
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Global variable to store extracted book text
global_book_text = ""

# Function to extract text from the uploaded PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with open(pdf_path, "rb") as file:
            reader = PyPDF2.PdfReader(file)
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text += page.extract_text()
    except Exception as e:
        return f"Error occurred while reading the file: {str(e)}"

    if not text:
        return "No text extracted from the book. Please ensure the file contains extractable text."

    return text

# Function to generate a question from a given text using T5
def generate_question_from_text(text):
    try:
        # Split the text into smaller chunks
        text_chunks = [text[i:i+1000] for i in range(0, len(text), 1000)]
        chunk = random.choice(text_chunks)  # Randomly select a chunk

        prompt = f"Generate a question based on the following text:\n\n{chunk}"

        # Tokenize the input text
        inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True)
        outputs = model.generate(**inputs, max_length=50)  # Use max_length to control question length

        # Decode the generated output
        question = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Ensure the question length is 8 words
        words = question.split()
        if len(words) > 8:
            question = ' '.join(words[:8]) + '?'
        elif not question.endswith('?'):
            question += '?'

    except Exception as e:
        return f"Error occurred while generating the question: {str(e)}"

    return question

# Function to generate three questions from the PDF
def generate_three_questions_from_pdf(pdf_file):
    global global_book_text
    # Extract text from the uploaded PDF
    book_text = extract_text_from_pdf(pdf_file.name)

    # If there is an error with the text, return the error directly
    if "Error occurred" in book_text:
        return book_text

    # Store extracted text globally for later use in evaluation
    global_book_text = book_text

    # Generate three questions from the book text
    questions = [generate_question_from_text(book_text) for _ in range(3)]

    return questions

# Function to calculate similarity between student's answer and book text
def evaluate_student_answer(student_answer, book_text):
    try:
        # Use TF-IDF to calculate similarity
        vectorizer = TfidfVectorizer()
        vectors = vectorizer.fit_transform([student_answer, book_text])

        # Calculate cosine similarity
        similarity = cosine_similarity(vectors[0], vectors[1])[0][0]

        # Convert similarity to percentage
        similarity_percentage = similarity * 100
    except Exception as e:
        return f"Error occurred while calculating similarity: {str(e)}"

    return similarity_percentage

# Function to evaluate answers for all three questions
def evaluate_all_answers(student_answers):
    global global_book_text
    scores = []

    # Evaluate each answer
    for answer in student_answers:
        score = evaluate_student_answer(answer, global_book_text)
        scores.append(score)

    # Calculate the total score
    total_score = sum(scores) / len(scores)

    return scores, total_score

# Function to save student's results to an Excel file
def save_results_to_excel(student_name, student_level, scores, total_score):
    file_path = "student_scores.xlsx"

    # If file exists, open it and append the data
    if os.path.exists(file_path):
        df = pd.read_excel(file_path)
    else:
        # If it doesn't exist, create a new file
        df = pd.DataFrame(columns=["Student Name", "Level", "Score 1", "Score 2", "Score 3", "Total Score"])

    # Add new data
    new_data = pd.DataFrame({
        "Student Name": [student_name],
        "Level": [student_level],
        "Score 1": [scores[0]],
        "Score 2": [scores[1]],
        "Score 3": [scores[2]],
        "Total Score": [total_score]
    })

    df = pd.concat([df, new_data], ignore_index=True)

    # Save data to Excel file
    df.to_excel(file_path, index=False)

# Gradio interface for step one: generating questions
def step_one(pdf_file):
    # Generate three questions from the book
    questions = generate_three_questions_from_pdf(pdf_file)

    # Return the questions
    return questions[0], questions[1], questions[2]

# Gradio interface for step two: evaluating answers and saving student name and score
def step_two(answer1, answer2, answer3, student_name, student_level):
    # Evaluate each answer individually
    scores, total = evaluate_all_answers([answer1, answer2, answer3])

    # Save student name, scores, and level in Excel file
    save_results_to_excel(student_name, student_level, scores, total)

    return (
        f"Evaluation for Question 1: {scores[0]:.2f}%",
        f"Evaluation for Question 2: {scores[1]:.2f}%",
        f"Evaluation for Question 3: {scores[2]:.2f}%",
        f"Total Score: {total:.2f}%",
        f"Student name: {student_name}, Level: {student_level}, and scores have been saved."
    )

# Gradio interface for generating questions and evaluating answers in two steps
with gr.Blocks() as demo:
    # Step 1: Upload the book and generate questions
    with gr.Row():
        pdf_input = gr.File(label="Upload PDF Book")
        question1_output = gr.Textbox(label="Question 1", interactive=False)
        question2_output = gr.Textbox(label="Question 2", interactive=False)
        question3_output = gr.Textbox(label="Question 3", interactive=False)
        generate_question_button = gr.Button("Generate Questions")

    generate_question_button.click(step_one, inputs=pdf_input, outputs=[question1_output, question2_output, question3_output])

    # Step 2: Input student's answers and evaluate them
    with gr.Row():
        student_answer1_input = gr.Textbox(label="Student's Answer for Question 1")
        student_answer2_input = gr.Textbox(label="Student's Answer for Question 2")
        student_answer3_input = gr.Textbox(label="Student's Answer for Question 3")
        student_name_input = gr.Textbox(label="Student's Name")
        student_level_input = gr.Dropdown(
            label="Choose Your Academic Level",
            choices=["LEVEL 000", "LEVEL 100", "LEVEL 200", "LEVEL 300", "LEVEL 400"],
            value="LEVEL 000"
        )
        evaluation1_output = gr.Textbox(label="Evaluation for Question 1", interactive=False)
        evaluation2_output = gr.Textbox(label="Evaluation for Question 2", interactive=False)
        evaluation3_output = gr.Textbox(label="Evaluation for Question 3", interactive=False)
        total_output = gr.Textbox(label="Total Score", interactive=False)
        save_confirmation = gr.Textbox(label="Save Confirmation", interactive=False)
        evaluate_answer_button = gr.Button("Evaluate Answers and Save Results")

    evaluate_answer_button.click(step_two, inputs=[student_answer1_input, student_answer2_input, student_answer3_input, student_name_input, student_level_input], outputs=[evaluation1_output, evaluation2_output, evaluation3_output, total_output, save_confirmation])

# Launch Gradio interface
demo.launch(share=True)
```
